```python
import pandas as pd
```

```python
heart=pd.read_csv('/kaggle/input/heart-failure-prediction/heart.csv')
```

```python
len(heart)
```

```python
heart.head()
```

```python
i=0
for i in range(len(heart)):
    if heart['Sex'][i]=='F':
      heart['Sex'][i]=0
    elif heart['Sex'][i]=='M':
      heart['Sex'][i]=1
```

```python
i=0
for i in range(len(heart)):
    if heart['ChestPainType'][i]=='ASY':
      heart['ChestPainType'][i]=0
    elif heart['ChestPainType'][i]=='ATA':
      heart['ChestPainType'][i]=1
    elif heart['ChestPainType'][i]=='NAP':
      heart['ChestPainType'][i]=2
    elif heart['ChestPainType'][i]=='TA':
      heart['ChestPainType'][i]=3
    
```

```python
i=0
for i in range(len(heart)):
    if heart['RestingECG'][i]=='LVH':
      heart['RestingECG'][i]=0
    elif heart['RestingECG'][i]=='Normal':
      heart['RestingECG'][i]=1
    elif heart['RestingECG'][i]=='ST':
      heart['RestingECG'][i]=2    
```

```python
i=0
for i in range(len(heart)):
  if heart['ExerciseAngina'][i]=='Y':
    heart['ExerciseAngina'][i]=0
  elif heart['ExerciseAngina'][i]=='N':
    heart['ExerciseAngina'][i]=1
```

```python
i=0
for i in range(len(heart)):
    if heart['ST_Slope'][i]=='Down':
      heart['ST_Slope'][i]=0
    elif heart['ST_Slope'][i]=='Flat':
      heart['ST_Slope'][i]=1
    elif heart['ST_Slope'][i]=='Up':
      heart['ST_Slope'][i]=2    
```

```python
heart.head()
```

```python
from sklearn.model_selection import train_test_split
Train,Test = train_test_split(heart,shuffle=True)
```

```python
from sklearn.svm import SVC
classifier = SVC(C=1000, kernel = 'rbf',random_state = 0)
classifier.fit(Train[['Age','Sex','ChestPainType','RestingBP','Cholesterol','FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope']],Train['HeartDisease'])
```

```python
predictions=classifier.predict(Test[['Age','Sex','ChestPainType','RestingBP','Cholesterol','FastingBS','RestingECG','MaxHR','ExerciseAngina','Oldpeak','ST_Slope']])
predictions
```

```python
Test['HeartDisease']
```

```python
from sklearn.metrics import precision_score,accuracy_score,recall_score,f1_score,classification_report
precision = precision_score(Test["HeartDisease"], predictions)
accuracy = accuracy_score(Test["HeartDisease"], predictions)
recall=recall_score(Test["HeartDisease"], predictions)
f1=f1_score(Test["HeartDisease"],predictions)
ClassificationReport=classification_report(Test["HeartDisease"], predictions)
print("\nAccuracy: ",accuracy)
print("\nPrecision: ",precision)
print("\nRecall: ",recall)
print("\nF1 Score: ",f1)
print("\nClassification Report: \n")
print(ClassificationReport)
```
